Q2. Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy. Compare the timing of both solutions. Which one is faster? Why?

```python
import numpy as np
import time
import matplotlib.pyplot as plt
n = 4
m = 3
k = 2

a1 = np.random.rand(n, k)
a2 = np.random.rand(k, m)

result = np.zeros((n, m))

start_time = time.time()
for i in range(0, n):
    for j in range(0, m):
        for l in range(0, k):
            result[i][j] += a1[i][l] * a2[l][j]
manual_multiplication_time = time.time() - start_time

start_time = time.time()
np_result = np.matmul(a1, a2)
numpy_multiplication_time = time.time() - start_time


print("Time for manual matrix multiplication:", manual_multiplication_time)
print("Time for NumPy's matrix multiplication:", numpy_multiplication_time)
```

For a list of size n, the median would be the ([n/2]+1)th largest element for odd n and the average of (n/2)th and (n/2 +1)th element for even n. Then, the time complexity required is $O(n^2)$.

Better method is merge sort where time complexity is $O(n\log(n))$

```python
import numpy as np
import time

arr = [1,2,4,6,3,8]
sorted_arr = []

start_time = time.time()


def sort(arr):
    for i in range(0,len(arr)):
        highest = 0
        for j in range(0,len(arr)):
            if arr[j] > highest:
                highest = arr[j]
                highest_index = j
        sorted_arr.append(highest)
        arr[highest_index] = 0
    return sorted_arr

sort(arr)
n = len(arr)        
if(n%2 == 0):
    median_slow = (sorted_arr[n//2] +sorted_arr[n//2 + 1])/2     
else:
    median_slow = sorted_arr[(n+1)/2]

print("Time for slow median is",time.time()-start_time)

np.start = time.time()
np.median(arr)

print("Time for numpy median is",time.time()-np.start)


                
```

Q4. What is the gradient of the following function with respect to x and y?


```python
from sympy import symbols,diff,sin

x,y = symbols('x y')
f = x**2 * y + y**3 * sin(x)

wrtx = diff(f,x)
wrty = diff(f,y)
print("gradient wrt to x is ", wrtx)
print("gradient wrt to y is", wrty)
```

Q5. Use JAX to confirm the gradient evaluated by your method matches the analytical solution corresponding to a few random values of x and y

```python
import jax
import jax.numpy as jnp
import math
import numpy as np


def custom_expression(x, y):
    return x**2 * y + y**3 * jnp.sin(x)

def custom_gradient(x_val, y_val):
    gradient_x = 2 * x_val * y_val + y_val**3 * math.cos(x_val)
    gradient_y = x_val**2 + 3 * y_val**2 * jnp.sin(x_val)
    return jnp.array([gradient_x, gradient_y])

gradient_func = jax.grad(custom_expression, argnums=(0, 1))

x_val, y_val = np.random.randn(), np.random.randn()

print(jnp.array(gradient_func(x_val, y_val)))
print(custom_gradient(x_val, y_val))
```

Q8. Store the same information using Python classes. We have an overall database which is a list of year objects. Each year contains a list of branches. Each branch contains a list of students. Each student has some properties like name, roll number and has marks in some subjects.

```python
class AcademicYear:
    def __init__(self, year):
        self.year = year
        self.branches = []

class Branch:
    def __init__(self, branch):
        self.branch = branch
        self.students = []

class Student:
    def __init__(self, name, roll_number):
        self.name = name
        self.roll_number = roll_number
        self.subjects = []

class Subject:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

academic_batch = AcademicYear('2022')
branch_one = Branch('Branch 1')
student_n1 = Student('n', '1')
math_subject = Subject('Mathematics', 100)
english_subject = Subject('English', 70)

academic_batch.branches.append(branch_one)
branch_one.students.append(student_n1)
student_n1.subjects.extend([math_subject, english_subject])
```

Name the columns of the dataframe as “a”, “b”, “c”, “d”, “e”. Find the column with the highest standard deviation. Find the row with the lowest mean.

Q11 Add a new column to the dataframe called “f” which is the sum of the columns “a”, “b”, “c”, “d”, “e”. Create another column called “g”. The value in the column “g” should be “LT8” if the value in the column “f” is less than 8 and “GT8” otherwise. Find the number of rows in the dataframe where the value in the column “g” is “LT8”. Find the standard deviation of the column “f” for the rows where the value in the column “g” is “LT8” and “GT8” respectively.

```python
import numpy as np
import pandas as pd

matrix = np.random.uniform(1,2, size=(20,5))
df = pd.DataFrame(matrix,columns=["a","b","c","d","e"])

mean_df = df.mean(axis=0)
std_df = df.std(axis=1)

print("Column with highest mean is",mean_df.idxmax())
print("Row with lowest standard deviation is",std_df.idxmin())


df['f'] = df.sum(axis=1)
df['g'] = df['f'].apply(lambda x:'LT8' if x<8 else 'GT8')
print("No. of rows with value LT8 in g:",len(df[df['g'] == 'LT8']))
print("For value LT8, standard deviation of f:",df[df['g'] == 'LT8']['f'].std())
print("For value GT8, standard deviation of f:",df[df['g'] == 'GT8']['f'].std())
```

Q13. Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.

```python
import numpy as np

def argmin(arr):
    min = arr[0]
    x = 0
    for i in range (len(arr)):
        if min > arr[i]:
            min = arr[i]
            x = i
    return x    
a = np.array([24,62,4,2,6,2,7,9,5,3,7,4,643,2,32,1,55,3,22,5])
print(argmin(a))
print(np.argmin(a))
```
